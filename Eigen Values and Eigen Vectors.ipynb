```python
#import all libraries
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#Load Dataset
read=pd.read_csv('temp_data.csv')
print("Dataset is:- ")
read.head()
```

```python
matrix = read.to_numpy()
print("Given Matrix is:")
print(matrix)
```

```python
#find transpose of given matrix
print("Transpose of a Given Matrix")
trans = [[matrix[j][i] for j in range(len(matrix))] for i in range(len(matrix[0]))]
for transpose in trans:
   print(transpose)
```

```python
#find AA'
print("value of AA' is:- ")
dot_pro = [[sum(first*second for first,second in zip(X_rows,Y_column)) for Y_column in zip(*trans)] for X_rows in matrix]
for dot in dot_pro:
   print(dot)
```

```python
eigenvalues, eigenvectors = np.linalg.eigh(dot_pro)

print("Eigen Values is:- ")
print(eigenvalues)
```

```python
print("Eigen Vectors is:- ")
print(eigenvectors)
```
